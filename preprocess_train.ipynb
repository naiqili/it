```python
time_dict = {}

def time_to_id(time):
    if time not in time_dict.keys():
        time_dict[time] = len(time_dict.keys())
    return time_dict[time]
```

```python
with open(train_file) as f:
    l = f.readline()
    print(l)
    for l in f:
        #print(l.strip().split(','))
        l = l.strip().split(',')
        time_to_id(l[2] + '-' + l[3])
        
```

```python
all_time = list(time_dict.keys())
all_time.sort()
print(len(all_time))
print(all_time)
```

```python
time_dict = {}

for i in range(len(all_time)):
    time_dict[all_time[i]] = i
```

```python
np_data = np.ones((112547, 512))*-1

with open(train_file) as f:
    l = f.readline()
    for l in f:
        #print(l.strip().split(','))
        l = l.strip().split(',')
        _id = stock_to_id(int(l[1]))
        _t = time_to_id(l[2]+'-'+l[3])
        _p = float(l[-1])
        np_data[_t, _id] = _p
```

```python
print(np.where(np_data < 0))
```

```python
for j in range(512):
    if np_data[0, j] < 0:
        k = np.min(np.where(np_data[:, j] > 0))
        #print(j, k)
        #break
        np_data[0, j] = np_data[k, j]
        
for i in range(1, 112547):
    for j in range(512):
        if np_data[i, j] < 0:
            np_data[i, j] = np_data[i-1, j]
```

```python
import pickle

stock_dict_file = './tmp/stock_dict.pkl'
time_dict_file = './tmp/time_dict.pkl'
np_data_file  = './tmp/data.npz'

with open(stock_dict_file, 'wb') as f:
    pickle.dump(stock_dict, f)

with open(time_dict_file, 'wb') as f:
    pickle.dump(time_dict, f)

with open(np_data_file, 'wb') as f:
    np.save(f, np_data)
```
